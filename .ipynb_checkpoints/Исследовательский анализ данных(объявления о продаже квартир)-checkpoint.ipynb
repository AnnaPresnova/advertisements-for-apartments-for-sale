## Предобработка данных

```python
# заполним пропуски в колонке - количество балконов и поменяем тип на int

df['balcony'] = df['balcony'].fillna(0)
df['balcony'] = df['balcony'].astype('int')

# поменяем тип данных в колонке - дата публикации

df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')

# посмотрим первые 20 строк датафрейма, проверим, как заполнились пропуски в колонке - количество балконов

display(df.head(20))

# посмотрим информацию о датафрейме, проверим изменение форматов

df.info()
```

```python
# создадим функцию, которая заменит тип float64 на float32:

def type_change(columns):
    df[columns] = df[columns].astype('float32')
    return df[columns]

type_change('total_area')
type_change('living_area')
type_change('kitchen_area')

# проверим, изменился ли тип:

df.info()
```

* высота потолков - пропущенные значения не повлияют на дальнейшие расчеты, оставим незаполненными,

## Расчёты и добавление результатов в таблицу

```python
# добавляем в таблицу цену квадратного метра

df['price_sq_m'] = (df['last_price'] / df['total_area']).astype('int')

# добавляем в таблицу день недели, месяц и год публикации объявления

df['weekday'] = df['first_day_exposition'].dt.weekday
df['month_exposition'] = df['first_day_exposition'].dt.month
df['year_exposition'] = df['first_day_exposition'].dt.year

# с помощью функции добавляем в таблицу этаж квартиры (варианты — первый, последний, другой)

def floor_variant(row):
    floors_total = row['floors_total'] 
    floor = row['floor']
    if floor == 1:
        return 'первый'
    if floor == floors_total:
        return 'последний'
    return 'другой'
    
df['floor_variant'] = df.apply(floor_variant, axis=1)

# добавляем в таблицу соотношение жилой и общей площади, а также отношение площади кухни к общей.

df['area_ratio'] = (df['living_area'] / df['total_area']).round(2)
df['kitchen_area_ratio'] = (df['kitchen_area'] / df['total_area']).round(2)

#выводим первые 20 строк таблицы

display(df.head(20))

# посмотрим статистические данные в колонке с ценой квадратного метра

df['price_sq_m'].describe().round(2)
```

### Вывод

Посчитаны и добавлены в таблицу:
* цена квадратного метра;
* день недели, месяц и год публикации объявления;
* этаж квартиры; варианты — первый, последний, другой;
* соотношение жилой и общей площади, а также отношение площади кухни к общей.

Цену квадратного метра перевели в int, а соотношения площадей округлили до двух знаков после запятой.
При выводе первых 20 строк датафрейма была выявлена очень высокая цена за квадратный метр (более 408 тыс руб за кв м), поэтому проанализируем статистические данные по этой колонке:
* **аномально низкая минимальная цена за квадратный метр (111 руб)**,
* **аномально высокая максимальная цена за квадратный метр (1 907 500 руб)**.

Проанализируем построенный график и опишем каждый фактор:

* общая площадь: прослеживается небольшая зависимость - чем больше площадь, тем выше цена; квартир с большими площадьми больше в центре города.

```python
# построим графики зависимости цены от этажа и от даты размещения объявления

df_good.pivot_table(index='floor_variant', values='last_price', aggfunc='median').plot(grid=True, figsize=(12,5), title='Зависимость цены от этажа')

df_good.pivot_table(index='weekday', values='last_price', aggfunc='median').plot(grid=True, figsize=(12,5), title='Зависимость цены от дня недели')

df_good.pivot_table(index='month_exposition', values='last_price', aggfunc='median').plot(grid=True, figsize=(12,5), title='Зависимость цены от месяца')

df_good.pivot_table(index='year_exposition', values='last_price', aggfunc='median').plot(grid=True, figsize=(12,5), title='Зависимость цены от года')
```

#### Вывод

Проанализировав зависимость цены от этажа квартиры, делаем вывод, что самую низкую стоимость имеют квартиры, расположенные на первом этаже, а самую высокую - квартиры на другом этаже (не на первом и не на последнем).
Зависимость цены от дня недели наблюдается следующая: самая высокая цена во вторник, самая низкая  - в воскресенье.
Зависимость цены от месяца: самая высокая цена - в мае и сентябре, самая низкая - в марте и июне.
Зависимость цены от года: самая высокая цена - в 2014 году, самая низкая  - в 2017 году. В 2014 году рост обусловлен экономическим кризисом, после чего цена стала падать до 2017 года, далее к 2019 году наблюдается рост.

## Общий вывод

Исходя из результатов исследований, отмечаем, что значимым параметром, который оказывает влияние на цену квартиры является площадь квартиры (количество комнат тоже оказывает влияние, но меньшее). Менее значимым параметром является удаленность от центра.
